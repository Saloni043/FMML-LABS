### Gradient Descent

````python
# Define the derivative function df(x) and the original function f(x) if not defined
def df(x):
    # Example: Derivative of f(x) = x^2 is df(x) = 2x
    return 2 * x

def f(x):
    # Example: f(x) = x^2
    return x ** 2

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED]
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x) and its derivative dg(x)
def g(x):
    return x**2 + 3  # Example: A quadratic function

def dg(x):
    return 2 * x  # Derivative of g(x)

# Define the gradient descent function with max iterations
def gradient_descent(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=500):
    """
    Perform gradient descent to find the local minimum.

    Parameters:
        derivative_func: function, derivative of f(x)
        initial_guess: float, initial value of x
        multiplier: float, step size or learning rate
        precision: float, minimum step size to stop descent
        max_iter: int, maximum number of iterations

    Returns:
        local_min: float, x value at local minimum
        list_x: list, history of x values during descent
        deriv_list: list, history of derivative values
    """
    current_x = initial_guess
    list_x = [current_x]
    deriv_list = [derivative_func(current_x)]

    for _ in range(max_iter):
        gradient = derivative_func(current_x)
        next_x = current_x - multiplier * gradient

        list_x.append(next_x)
        deriv_list.append(derivative_func(next_x))

        if abs(next_x - current_x) < precision:
            break

        current_x = next_x

    return current_x, list_x, deriv_list

# Call the gradient descent function
local_min, list_x, deriv_list = gradient_descent(
    derivative_func=dg,
    initial_guess=1.9,
    multiplier=0.02,
    max_iter=500
)

# Generate x values for plotting
x_2 = np.linspace(-2, 2, 100)

# Plot function and derivative and scatter plot side by side
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)
plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)  # Plot g(x)
plt.scatter(list_x, g(np.array(list_x)), color='red', s=100, alpha=0.6)  # Scatter gradient descent points

# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)  # Plot dg(x)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)  # Scatter derivative points

plt.show()

# Print number of steps
print('Number of steps is:', len(list_x))
````

````python
# Run gradient descent with different learning rates
n = 100

# Low learning rate
low_gamma = gradient_descent(
    derivative_func=dg,
    initial_guess=3,
    multiplier=0.0005,
    precision=0.0001,
    max_iter=n
)

# Medium learning rate (mid_gamma)
mid_gamma = gradient_descent(
    derivative_func=dg,
    initial_guess=3,
    multiplier=0.001,  # Mid-level learning rate
    precision=0.0001,
    max_iter=n
)

# High learning rate (high_gamma)
high_gamma = gradient_descent(
    derivative_func=dg,
    initial_guess=3,
    multiplier=0.002,  # Higher learning rate
    precision=0.0001,
    max_iter=n
)

# Insane learning rate
insane_gamma = gradient_descent(
    derivative_func=dg,
    initial_guess=3,
    multiplier=0.01,  # Very high learning rate
    precision=0.0001,
    max_iter=n
)

# Plot results for low, mid, and high learning rates
plt.figure(figsize=[15, 10])

# Plot Low Gamma
plt.subplot(2, 2, 1)
plt.title("Low Gamma (0.0005)", fontsize=17)
plt.xlabel("Steps", fontsize=14)
plt.ylabel("g(x)", fontsize=14)
plt.plot(low_gamma[1], g(np.array(low_gamma[1])), color='blue', label='Low Gamma')
plt.scatter(range(len(low_gamma[1])), g(np.array(low_gamma[1])), color='red', alpha=0.6)
plt.legend()

# Plot Mid Gamma
plt.subplot(2, 2, 2)
plt.title("Mid Gamma (0.001)", fontsize=17)
plt.xlabel("Steps", fontsize=14)
plt.ylabel("g(x)", fontsize=14)
plt.plot(mid_gamma[1], g(np.array(mid_gamma[1])), color='green', label='Mid Gamma')
plt.scatter(range(len(mid_gamma[1])), g(np.array(mid_gamma[1])), color='red', alpha=0.6)
plt.legend()

# Plot High Gamma
plt.subplot(2, 2, 3)
plt.title("High Gamma (0.002)", fontsize=17)
plt.xlabel("Steps", fontsize=14)
plt.ylabel("g(x)", fontsize=14)
plt.plot(high_gamma[1], g(np.array(high_gamma[1])), color='orange', label='High Gamma')
plt.scatter(range(len(high_gamma[1])), g(np.array(high_gamma[1])), color='red', alpha=0.6)
plt.legend()

# Plot Insane Gamma
plt.subplot(2, 2, 4)
plt.title("Insane Gamma (0.01)", fontsize=17)
plt.xlabel("Steps", fontsize=14)
plt.ylabel("g(x)", fontsize=14)
plt.plot(insane_gamma[1], g(np.array(insane_gamma[1])), color='purple', label='Insane Gamma')
plt.scatter(range(len(insane_gamma[1])), g(np.array(insane_gamma[1])), color='red', alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Example data (x, y) for a simple linear regression task
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 1.3, 3.75, 2.25])

# Define the hypothesis function (linear model)
def hypothesis(theta0, theta1, x):
    return theta0 + theta1 * x

# Define the Mean Squared Error (MSE) cost function
def cost_function(theta0, theta1, x, y):
    m = len(y)  # Number of training examples
    predictions = hypothesis(theta0, theta1, x)
    return (1/(2*m)) * np.sum((predictions - y)**2)

# Create a grid for theta0 and theta1
theta0_vals = np.linspace(-10, 10, 100)  # Range for theta0
theta1_vals = np.linspace(-10, 10, 100)  # Range for theta1
plot_t0, plot_t1 = np.meshgrid(theta0_vals, theta1_vals)  # Create a grid of theta values

# Calculate the cost (MSE) for each pair of theta0 and theta1 values
plot_cost = np.zeros_like(plot_t0)
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        plot_cost[j, i] = cost_function(plot_t0[j, i], plot_t1[j, i], x, y)

# Plotting the MSE surface
fig = plt.figure(figsize=[16, 12])
ax = fig.add_subplot(111, projection='3d')

# Labels
ax.set_xlabel('Theta 0', fontsize=20)
ax.set_ylabel('Theta 1', fontsize=20)
ax.set_zlabel('Cost - MSE', fontsize=20)

# Plot the surface
ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap=cm.hot)

# Show the plot
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````
